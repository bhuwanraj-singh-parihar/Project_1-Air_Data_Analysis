```python
pollutant_summary = (
    air_data
    .groupby('sensor.name')
    .agg(
        mean_voc=('voc', 'mean'),
        median_voc=('voc', 'median'),
        mean_pm25=('pm2.5_atm', 'mean'),
        median_pm25=('pm2.5_atm', 'median'),
        mean_pm10=('pm10.0_atm', 'mean'),
        median_pm10=('pm10.0_atm', 'median')
    )
    .reset_index()
)
```

```python
top5_mean_voc = pollutant_summary.sort_values('mean_voc', ascending=False).head(5)
top5_mean_voc
```

```python
air_data.groupby('temperature')['pm2.5_atm'].mean()
```

```python
def assign_temp_category(t):
    if t < 32:
        return 'Cold'
    elif 32 <= t <= 50:
        return 'Cool'
    elif 51 <= t <= 70:
        return 'Mild'
    else:
        return 'Warm/Hot'

air_data['temp_category'] = air_data['temperature'].apply(assign_temp_category)

air_data[['temperature', 'temp_category']].head()
```

```python
air_data.groupby('temp_category')['pm2.5_atm'].mean()
```

```python
def assign_humidity_category(h):
    if h < 30:
        return 'Low'
    elif 30 <= h <= 60:
        return 'Moderate'
    else:
        return 'High'

air_data['humidity_category'] = air_data['humidity'].apply(assign_humidity_category)
air_data.groupby('humidity_category')['pm2.5_atm'].mean()
```

```python
def assign_aqi_category(pm):
    if pm <= 12:
        return "Good"
    elif pm <= 35.4:
        return "Moderate"
    elif pm <= 55.4:
        return "Unhealthy for Sensitive Groups"
    elif pm <= 150.4:
        return "Unhealthy"
    else:
        return "Very Unhealthy/Hazardous"

air_data['pm25_aqi_category'] = air_data['pm2.5_atm'].apply(assign_aqi_category)
air_data['pm25_aqi_category'].value_counts()
```

```python
def assign_altitude(y):
    if y < 2500:
        return 'Low'
    elif 2500 <= y <= 3000:
        return 'Medium'
    else:
        return 'High'

air_data['altitude_category'] = air_data['sensor.altitude'].apply(assign_altitude)

air_data[['sensor.altitude', 'altitude_category']].head()
```

```python
air_data.groupby('altitude_category')['pm2.5_atm'].mean()
```
